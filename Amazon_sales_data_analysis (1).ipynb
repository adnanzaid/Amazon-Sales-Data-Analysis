How many orders are associated with promotions (promotion-ids) versus those without?


```python
Orders_with_promotion = (data['promotion-ids'].notnull() & data['promotion-ids'].str.strip().ne(""))

# Orders with promotion

with_promotion = (Orders_with_promotion).sum()
print(with_promotion)

# Orders without promotion
without_promotion = (~Orders_with_promotion).sum()
print(without_promotion)
```

What is the total sales (Amount) for B2B (B2B = Yes) orders compared to non-B2B orders?


```python
# B2B Sales
b2b_sales = data[data['B2B'] == True ]['Amount'].sum()
print(b2b_sales)

# B2B Sales
non_b2b_sales = data[data['B2B'] == False]['Amount'].sum()
print(non_b2b_sales)


```
